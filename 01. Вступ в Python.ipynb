# 1. Встпу в Python

## 1.1 Python  і його переваги

**Python** (найчастіше вживане прочитання — «Па́йтон», запозичено назву з британського шоу Монті Пайтон) — інтерпретована об'єктно-орієнтована мова програмування високого рівня зі строгою динамічною типізацією. 

Розроблена в 1990 році Гвідо ван Россумом. 

Структури даних високого рівня разом із динамічною семантикою та динамічним зв'язуванням роблять її привабливою для швидкої розробки програм, а також як засіб поєднування наявних компонентів. Python підтримує модулі та пакети модулів, що сприяє модульності та повторному використанню коду. Інтерпретатор Python та стандартні бібліотеки доступні як у скомпільованій, так і у вихідній формі на всіх основних платформах. В мові програмування Python підтримується кілька парадигм програмування, зокрема: об'єктно-орієнтована, процедурна, функціональна та аспектно-орієнтована.

#### Переваги Python:
* чистий синтаксис (для виділення блоків слід використовувати відступи);
* переносність програм (що властиве більшості інтерпретованих мов);
* стандартний дистрибутив має велику кількість корисних модулів (включно з модулем для розробки графічного інтерфейсу);
* можливість використання Python в діалоговому режимі (дуже корисне для експериментування та розв'язання простих задач);
* стандартний дистрибутив має просте, але разом із тим досить потужне середовище розробки, яке зветься IDLE і яке написане мовою Python;
* зручний для розв'язання математичних проблем (має засоби роботи з комплексними числами, може оперувати з цілими числами довільної величини, у діалоговому режимі може використовуватися як потужний калькулятор);
* відкритий код (можливість редагувати його іншими користувачами).

## 1.2 Встановлення Python

У багатьох операційних системах Python встановлений за замовчуванням. 

Але наша задача встановити найновішу версію мови та перевірити, чи коректо все працює

- *Windows*

Для даної ОС потрбіно скачати установщик (<a href = 'https://www.python.org/downloads/'>натисніть тут</a>) та запустити його.

Обов'язково поставте галочку для **Add Python 3.8 to PATH**, це дасть можливість запускати код  з консолі. Далі просто слідуйте інструкції

- *Mac OS*

Аналоічно як і в Windows, скачайте установщик для Mac OS (<a href = 'https://www.python.org/downloads/'>натисніть тут</a>) та запустіть його, після - запустіть *Python.mpkg*

- *Linux*

В різних дистрибутивах встановленння може відрізнятися, проте ви можете зібрати Python із вихідного коду.

Наприклад, в *Ububtu* встановити програму дуже просто:

<code>sudo apt-get update

sudo apt-get install python3.6</code>

## 1.3 Встановлення віртуального середовища

Що взагалі таке віртуальне середовище? Які його соновні задачі?

Інколи буває, що одна програма написана на старій версіії Python, інша - на новій. Чи, різні проекти, над якими ви працюєте, підтримують одну бібліотеку (один модуль), але теж, різних релізів. 

Було б добре якось ізолювати проекти один від одного. В цьому нам допоможе *віртуальне середовище*

Для Python найчастіше використовують virtualenv. Для його встановлення в консолі потрібно ввести:

<code>pip install virtualenv</code>

Щоб створити ізольоване середовище введіть:

<code>python3 -m venv name_env</code>

Для активації середовища у 
- *Windows*:

<code>source name_env/Scripts/activate</code>

- *Linux*:

<code>source name_env/Scripts/activate</code>

Для декавтивації використовуємо команду <code>deactivate</code>



Більш детальна інформація тут: https://virtualenv.pypa.io/
